```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Data Analytics Course\My Portfolio Projects\Project3-World Happiness Records in Python\2019_cleaned.csv")
df
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Top 10 Happiest Countries in 2019

top10 = df.sort_values('Happiness_Score', ascending=False).head(10)
sns.barplot(x='Happiness_Score', y='Country_or_Region', data=top10)
plt.title('Top 10 Happiest Countries (2019)')
plt.show()
```

```python
# Which factors most strongly correlate with happiness?
numeric_df = df.select_dtypes(include=['number'])
corr_happiness = numeric_df.corr()['Happiness_Score'].drop(['Happiness_Score'], errors='ignore')

plt.figure(figsize=(8, 5))
sns.heatmap(corr_happiness.to_frame(), annot=True, cmap='coolwarm', cbar=False)
plt.title('Happiness Correlates')
plt.show()
```

```python
# The Relationship Between Happiness and Its Strongest Correlate
sns.scatterplot(x='Economy(GDP_per_Capita)', y='Happiness_Score', data=df)
plt.title('Economy vs. Happiness Score')
plt.show()
```

```python
# The Relationship Between Happiness and Its Weakest Correlate
sns.scatterplot(x='Generosity', y='Happiness_Score', data=df)
plt.title('Generosity vs. Happiness Score')
plt.show()
```

```python
all_sheets = pd.read_excel(r"D:\Data Analytics Course\My Portfolio Projects\Project3-World Happiness Records in Python\World Happiness Report-cleaned dataset.xlsx",
    sheet_name=None)
df2 = pd.concat(all_sheets, ignore_index=False)
df2.reset_index(inplace=True)
df2.rename(columns={'level_0': 'Year'}, inplace=True)
df2.drop(columns=['level_1'], inplace=True)
df2
```

```python
# Average Happiness per Year

avg_year = df2.groupby('Year')['Happiness_Score'].mean().reset_index()
avg_year
```

```python
# Average Happiness per Year Visualization

avg_year = df2.groupby('Year')['Happiness_Score'].mean().reset_index()
sns.lineplot(x='Year', y='Happiness_Score', data=avg_year)
plt.title('Average Global Happiness Over Time')
plt.show()
```

```python
# Iran Happiness Report

Iran = df2[df2['Country'] == 'Iran']
Iran = Iran.reset_index(drop=True)
Iran
```

```python
factors = ['Happiness_Score', 'Economy(GDP_per_Capita)', 'Social_Support', 'Health_(Life_Expectancy)', 'Freedom', 'Perception_of_Corruption', 'Generosity']
```

```python
melted = pd.melt(
    Iran,
    id_vars='Year',
    value_vars=factors,
    var_name='Factors',
    value_name='Scores'
)

plt.figure(figsize=(10, 6))
sns.lineplot(data=melted, x='Year', y='Scores', hue='Factors', marker='o')
plt.title('Trends Across Years in Iran')
plt.xlabel('Year')
plt.ylabel('Scores')
plt.grid(False)
plt.show()
```
